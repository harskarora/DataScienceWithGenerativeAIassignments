````python
import numpy as np

# Generate a list of 100 integers between 90 and 130
int_list = np.random.randint(90, 131, size=100)

# Function to calculate the mean of a list of numbers
def mean(numbers):
    return np.mean(numbers)

# Function to calculate the median of a list of numbers
def median(numbers):
    return np.median(numbers)

# Program to compute the mode of a list of integers
def mode(numbers):
    return list(np.unique(numbers[np.argmax(np.apply_along_axis(np.bincount, 0, numbers, minlength=np.max(numbers) + 1), axis=0)]))

# Function to calculate the weighted mean of a list of values and their corresponding weights
def weighted_mean(values, weights):
    return np.average(values, weights=weights)

# Function to calculate the geometric mean of a list of positive numbers
def geometric_mean(numbers):
    return np.prod(numbers) ** (1 / len(numbers))

# Program to calculate the harmonic mean of a list of values
def harmonic_mean(numbers):
    return len(numbers) / np.sum(1 / numbers)

# Function to determine the midrange of a list of numbers
def midrange(numbers):
    return (np.min(numbers) + np.max(numbers)) / 2

# Program to find the trimmed mean of a list, excluding a certain percentage of outliers
def trimmed_mean(numbers, percentage):
    lower_percentile = np.percentile(numbers, percentage / 2)
    upper_percentile = np.percentile(numbers, 100 - percentage / 2)
    trimmed_numbers = numbers[(numbers >= lower_percentile) & (numbers <= upper_percentile)]
    return np.mean(trimmed_numbers)

print("Mean:", mean(int_list))
print("Median:", median(int_list))
print("Mode:", mode(int_list))
print("Weighted Mean:", weighted_mean(int_list, np.random.randint(1, 11, size=100)))
print("Geometric Mean:", geometric_mean(int_list))
print("Harmonic Mean:", harmonic_mean(int_list))
print("Midrange:", midrange(int_list))
print("Trimmed Mean:", trimmed_mean(int_list, 10))
````

````python
# Question 3: Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        mean = self.expected_value()
        return sum((value - mean) ** 2 * prob for value, prob in zip(self.values, self.probabilities))

values = [1, 2, 3, 4]
probabilities = [0.1, 0.2, 0.3, 0.4]
random_variable = DiscreteRandomVariable(values, probabilities)
print("Expected Value:", random_variable.expected_value())
print("Variance:", random_variable.variance())
````

````python
# Question 4: Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.

import random

class FairSixSidedDie:
    def __init__(self):
        self.sides = 6

    def roll(self):
        return random.randint(1, self.sides)

    def expected_value(self):
        return (1 + self.sides) / 2

    def variance(self):
        return (self.sides ** 2 - 1) / 12

num_rolls = 100000
die = FairSixSidedDie()
rolls = [die.roll() for _ in range(num_rolls)]

# Calculating the empirical mean and variance
mean = sum(rolls) / num_rolls
variance = sum((roll - mean) ** 2 for roll in rolls) / num_rolls

print("Empirical Expected Value:", mean)
print("Empirical Variance:", variance)
print("Theoretical Expected Value:", die.expected_value())
print("Theoretical Variance:", die.variance())
````

````python
# Question 5: Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.

import numpy as np

def generate_samples(distribution, params, size):
    if distribution == 'binomial':
        samples = np.random.binomial(n=params['n'], p=params['p'], size=size)
    elif distribution == 'poisson':
        samples = np.random.poisson(lam=params['lam'], size=size)
    else:
        raise ValueError("Unsupported distribution")

    mean = np.mean(samples)
    variance = np.var(samples)

    return samples, mean, variance

distribution = 'binomial'
params = {'n': 10, 'p': 0.5}  # parameters for binomial distribution
size = 1000  # number of samples
samples, mean, variance = generate_samples(distribution, params, size)

print("Generated Samples:", samples)
print("Mean:", mean)
print("Variance:", variance)
````

````python
#import numpy as np

def normal_pdf(x, mu, sigma):
    """
    Calculate the probability density function (PDF) of a normal distribution at a given point x.

    Parameters:
    - x (float or array-like): The point(s) at which to evaluate the PDF.
    - mu (float): The mean of the normal distribution.
    - sigma (float): The standard deviation of the normal distribution.

    Returns:
    - float or array-like: The value(s) of the PDF at the given point(s) x.
    """
    return 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-0.5 * ((x - mu) / sigma)**2)

mu = 0
sigma = 1
x = 1.5
pdf_value = normal_pdf(x, mu, sigma)
print("PDF at x =", x, ":", pdf_value)
````

````python
import numpy as np

def exponential_cdf(x, lambd):
    """
    Calculate the cumulative distribution function (CDF) of an exponential distribution at a given point x.

    Parameters:
    - x (float or array-like): The point(s) at which to evaluate the CDF.
    - lambd (float): The rate parameter of the exponential distribution (1/mean).

    Returns:
    - float or array-like: The value(s) of the CDF at the given point(s) x.
    """
    return 1 - np.exp(-lambd * x)

lambd = 0.5  # Rate parameter (1/mean)
x = 2.0
cdf_value = exponential_cdf(x, lambd)
print("CDF at x =", x, ":", cdf_value)
````

````python
import numpy as np
import math

def poisson_pmf(k, lambd):
    """
    Calculate the probability mass function (PMF) of a Poisson distribution at a given point k.

    Parameters:
    - k (int or array-like): The point(s) at which to evaluate the PMF.
    - lambd (float): The rate parameter of the Poisson distribution.

    Returns:
    - float or array-like: The value(s) of the PMF at the given point(s) k.
    """
    return np.exp(-lambd) * np.power(lambd, k) / np.array([math.factorial(int(x)) for x in k])

lambd = 2.5  # Rate parameter
k = [0, 1, 2, 3, 4]
pmf_values = poisson_pmf(k, lambd)
for i in range(len(k)):
    print("PMF at k =", k[i], ":", pmf_values[i])
````

````python
#A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors who make a purchase). They collect data from the old and new layouts to compare. To generate the data use the following command: ```python import numpy as np # 50 purchases out of 1000 visitors old_layout = np.array([1] * 50 + [0] * 950) # 70 purchases out of 1000 visitors  new_layout = np.array([1] * 70 + [0] * 930) ``` Apply z-test to find which layout is successful.

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Generate data
# 50 purchases out of 1000 visitors
old_layout = np.array([1] * 50 + [0] * 950)
# 70 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)

# Count the number of successes (purchases) and trials (total visitors) for both layouts
successes = np.array([old_layout.sum(), new_layout.sum()])
nobs = np.array([len(old_layout), len(new_layout)])

# Apply z-test
stat, p_value = proportions_ztest(successes, nobs)

print(f"Z-statistic: {stat}")
print(f"P-value: {p_value}")

# Determine which layout is more successful
if p_value < 0.05:  # Assuming a significance level of 0.05
    if stat > 0:
        print("The new layout leads to a higher conversion rate.")
    else:
        print("The old layout leads to a higher conversion rate.")
else:
    print("There is no significant difference in conversion rates between the old and new layouts.")
````

````python
#A tutoring service claims that its program improves students' exam scores. A sample of students who participated in the program was taken, and their scores before and after the program were recorded. Use the below code to generate samples of respective arrays of marks: ```python before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87]) after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88]) ``` Use z-test to find if the claims made by tutor are true or false.

import numpy as np
from scipy import stats

# Generate sample data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate the mean difference and standard error
mean_diff = np.mean(after_program - before_program)
std_diff = np.std(after_program - before_program, ddof=1) / np.sqrt(len(before_program))

# Calculate the z-score
z_score = mean_diff / std_diff

# Calculate the p-value for the z-score
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_score)))  # Two-tailed test

print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

# Determine if the claim is true or false
if p_value < 0.05:  # Assuming a significance level of 0.05
    print("The claim made by the tutor is true. The program improves students' exam scores.")
else:
    print("The claim made by the tutor is false. There is no significant improvement in students' exam scores.")
````

````python
#A company is testing two different website layouts to see which one leads to higher click-through rates. Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of freedom, and p-value. Use the following data: ```python layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37] layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

import numpy as np
from scipy import stats

def ab_test_analysis(data_a, data_b):
    """
    Perform an A/B test analysis to compare the means of two independent samples.
    
    Parameters:
    - data_a (list or array-like): Click-through rates for layout A.
    - data_b (list or array-like): Click-through rates for layout B.
    
    Returns:
    - t_stat (float): The t-statistic.
    - df (float): The degrees of freedom.
    - p_value (float): The p-value.
    """
    # Convert data to numpy arrays
    data_a = np.array(data_a)
    data_b = np.array(data_b)

    # Calculate the sample means
    mean_a = np.mean(data_a)
    mean_b = np.mean(data_b)

    # Calculate the sample standard deviations
    std_a = np.std(data_a, ddof=1)
    std_b = np.std(data_b, ddof=1)

    # Calculate the sample sizes
    n_a = len(data_a)
    n_b = len(data_b)

    # Calculate the pooled standard deviation
    pooled_std = np.sqrt(((std_a ** 2) / n_a) + ((std_b ** 2) / n_b))

    # Calculate the t-statistic
    t_stat = (mean_a - mean_b) / pooled_std

    # Calculate the degrees of freedom
    df = n_a + n_b - 2

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_stat), df))

    return t_stat, df, p_value

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

t_stat, df, p_value = ab_test_analysis(layout_a_clicks, layout_b_clicks)
print(f"T-statistic: {t_stat}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value}")

# Determine which layout is more successful
if p_value < 0.05:  # Assuming a significance level of 0.05
    if t_stat > 0:
        print("Layout A leads to higher click-through rates.")
    else:
        print("Layout B leads to higher click-through rates.")
else:
    print("There is no significant difference in click-through rates between the two layouts.")
````

````python
#A school district introduces an educational intervention program to improve math scores. Write a Python function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to determine if the intervention had a significant impact. Use the following data of test score: ```python pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87] post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

import numpy as np
from scipy import stats

def intervention_analysis(pre_scores, post_scores):
    """
    Analyze the impact of an educational intervention on test scores using a paired t-test.
    
    Parameters:
    - pre_scores (list or array-like): Test scores before the intervention.
    - post_scores (list or array-like): Test scores after the intervention.
    
    Returns:
    - t_stat (float): The t-statistic.
    - p_value (float): The p-value.
    """
    # Convert data to numpy arrays
    pre_scores = np.array(pre_scores)
    post_scores = np.array(post_scores)

    # Calculate the differences
    differences = post_scores - pre_scores

    # Calculate the mean and standard error of the differences
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)
    standard_error = std_diff / np.sqrt(len(differences))

    # Calculate the t-statistic
    t_stat = mean_diff / standard_error

    # Calculate the degrees of freedom
    df = len(differences) - 1

    # Calculate the p-value for the t-statistic (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_stat), df))

    return t_stat, p_value

# Example usage
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

t_stat, p_value = intervention_analysis(pre_intervention_scores, post_intervention_scores)
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Determine if the intervention had a significant impact
if p_value < 0.05:  # Assuming a significance level of 0.05
    print("The educational intervention had a significant impact on math scores.")
else:
    print("The educational intervention did not have a significant impact on math scores.")
````

````python
#An HR department wants to investigate if there's a gender-based salary gap within the company. Develop a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically significant difference between the average salaries of male and female employees. Use the below code to generate synthetic data: ```python # Generate synthetic salary data for male and female employees np.random.seed(0)  # For reproducibility male_salaries = np.random.normal(loc=50000, scale=10000, size=20) female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

import numpy as np
from scipy import stats

# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

def gender_salary_gap_analysis(male_salaries, female_salaries):
    """
    Analyze the gender-based salary gap using a t-test for independent samples.
    
    Parameters:
    - male_salaries (list or array-like): Salaries of male employees.
    - female_salaries (list or array-like): Salaries of female employees.
    
    Returns:
    - t_stat (float): The t-statistic.
    - p_value (float): The p-value.
    """
    # Convert data to numpy arrays
    male_salaries = np.array(male_salaries)
    female_salaries = np.array(female_salaries)

    # Calculate the sample means
    mean_male = np.mean(male_salaries)
    mean_female = np.mean(female_salaries)

    # Calculate the sample standard deviations
    std_male = np.std(male_salaries, ddof=1)
    std_female = np.std(female_salaries, ddof=1)

    # Calculate the sample sizes
    n_male = len(male_salaries)
    n_female = len(female_salaries)

    # Calculate the pooled standard deviation
    pooled_std = np.sqrt(((std_male ** 2) / n_male) + ((std_female ** 2) / n_female))

    # Calculate the t-statistic
    t_stat = (mean_male - mean_female) / pooled_std

    # Calculate the degrees of freedom
    df = n_male + n_female - 2

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_stat), df))

    return t_stat, p_value

# Analyze the salary data
t_stat, p_value = gender_salary_gap_analysis(male_salaries, female_salaries)
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Determine if there's a statistically significant difference
if p_value < 0.05:  # Assuming a significance level of 0.05
    print("There is a statistically significant difference between the average salaries of male and female employees.")
else:
    print("There is no statistically significant difference between the average salaries of male and female employees.")
````

````python
#A political analyst wants to determine if there is a significant association between age groups and voter preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify them into different age groups and candidate preferences. Perform a Chi-Square test to determine if there is a significant association between age groups and voter preferences. Use the below code to generate data: ```python np.random.seed(0) age_groups = np.random.choice(['18-30', '31-50', '51+', '51+'], size=30) voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

import numpy as np
from scipy import stats

def chi_square_test(age_groups, voter_preferences):
    """
    Perform a Chi-Square test for independence between age groups and voter preferences.
    
    Parameters:
    - age_groups (array-like): Array of age groups.
    - voter_preferences (array-like): Array of voter preferences.
    
    Returns:
    - chi2_stat (float): The Chi-Square statistic.
    - p_value (float): The p-value.
    - df (int): Degrees of freedom.
    - contingency_table (array): Contingency table of observed frequencies.
    """
    # Create a contingency table of observed frequencies
    contingency_table = np.zeros((3, 2))
    for i, age_group in enumerate(['18-30', '31-50', '51+']):
        for j, preference in enumerate(['Candidate A', 'Candidate B']):
            contingency_table[i, j] = np.sum((age_groups == age_group) & (voter_preferences == preference))

    # Perform the Chi-Square test
    chi2_stat, p_value, df, _ = stats.chi2_contingency(contingency_table)

    return chi2_stat, p_value, df, contingency_table

# Example usage
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

chi2_stat, p_value, df, contingency_table = chi_square_test(age_groups, voter_preferences)
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {df}")
print("Contingency Table:")
print(contingency_table)

# Determine if there is a significant association
if p_value < 0.05:  # Assuming a significance level of 0.05
    print("There is a significant association between age groups and voter preferences.")
else:
    print("There is no significant association between age groups and voter preferences.")
````
