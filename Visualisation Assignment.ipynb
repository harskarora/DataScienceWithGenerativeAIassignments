```python
# 4. Generate a dataset with categories and numerical values, and visualize the distribution.
data = pd.DataFrame({'Category': np.random.choice(['A', 'B', 'C', 'D', 'E'], 100),
                     'Value': np.random.normal(0, 1, 100)})
sns.boxplot(x='Category', y='Value', data=data)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Boxplot of Values across Categories')
plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features and visualize the correlation matrix using a heatmap.
data = pd.DataFrame(np.random.randn(100, 5), columns=['A', 'B', 'C', 'D', 'E'])
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 1. Generate a 3D scatter plot to visualize the distribution of data points in a three-dimensional space.
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
# 2. Create a violin plot to display the distribution of scores across different grade categories.
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)
sns.violinplot(x='Grade', y='Score', data=df)
plt.xlabel('Grade')
plt.ylabel('Score')
plt.title('Violin Plot of Scores across Grades')
plt.show()
```

```python
# 3. Generate a heatmap to visualize the variation in sales across different months and days.
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
pivot_df = df.pivot_table(index='Month', columns='Day', values='Sales', aggfunc=np.sum)
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_df, cmap='viridis', annot=True, fmt='g')
plt.xlabel('Day')
plt.ylabel('Month')
plt.title('Heatmap of Sales Variation across Months and Days')
plt.show()
```

```python
#4

# Given data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(df['X'], df['Y'], df['Z'], cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')
plt.show()
```

```python
#5 . Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (x-axis), and bubble size proportional to the population.

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Given dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Plotting the bubble chart
plt.figure(figsize=(12, 8))
plt.scatter(df['GDP'], df['Population'], s=df['Population']*0.5, alpha=0.7, c='skyblue', edgecolors='grey', linewidth=2)

# Adding labels and title
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart of Country GDP vs. Population')
plt.grid(True)
plt.xticks(np.arange(500, 2001, step=200))
plt.yticks(np.arange(100, 1001, step=100))
for i in range(len(df)):
    plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```

```python
#1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

from bokeh.plotting import figure, show
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="sin(x)", line_width=2)
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.


from bokeh.plotting import figure, show
import numpy as np

N = 100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.choice(["red", "green", "blue"], N)
sizes = np.random.randint(5, 20, N)

p = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y')
p.scatter(x, y, color=colors, size=sizes)
show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset. fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] counts = [20, 25, 30, 35]

from bokeh.plotting import figure, show
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts', plot_height=350)
p.vbar(x=fruits, top=counts, width=0.9)
show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data. data_hist = np.random.randn(1000) hist, edges = np.histogram(data_hist, bins=30)

from bokeh.plotting import figure, show
import numpy as np

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)
show(p)
```
